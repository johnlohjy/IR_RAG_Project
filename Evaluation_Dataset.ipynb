### Version 1 of Search functions

```python
# Function to get 1 positive example
def pos_search_query(query, query_number, search_term, search_columns, df, target_column):
    filtered_df = df[df[search_columns].str.contains(search_term, case=False, na=False)]
    filtered_df['word_count'] = filtered_df[target_column].str.split().apply(len)
    
    positive_df = filtered_df.copy()
    positive_df['query'] = query
    positive_df['query_number'] = query_number
    positive_df['sample'] = 'positive'

    print(filtered_df[[target_column, 'word_count']])
    print(f"length of content: {filtered_df[['word_count']]}")
    return positive_df
```

```python
# Function to get 3 random negative examples
def neg_search_query(query, query_number, search_term, search_columns, df, target_column):
    filtered_df = df[~df[search_columns].str.contains(search_term, case=False, na=False)]
    random_rows = filtered_df.sample(n=3, random_state=42)
    random_rows['word_count'] = filtered_df[target_column].str.split().apply(len)

    negative_df = random_rows.copy()
    negative_df['query'] = query
    negative_df['query_number'] = query_number
    negative_df['sample'] = 'negative'

    print(random_rows[['Article Links', target_column, 'word_count']])
    print(f"length of content: {random_rows[['word_count']]}")
    return negative_df
```

### Version 2: Improved search functions

```python
# Function to get 1 positive example, allowing for >1 search term using ['...', '...']
def improved_pos_search_query(query, query_number, search_terms, search_columns, df, target_column):
    pattern = '|'.join(search_terms)
    filtered_df = df[df[search_columns].str.contains(pattern, case=False, na=False)]
    filtered_df['word_count'] = filtered_df[target_column].str.split().apply(len)
    
    positive_df = filtered_df.copy()
    positive_df['query'] = query
    positive_df['query_number'] = query_number
    positive_df['sample'] = 'positive'

    print(filtered_df[[target_column, 'word_count']])
    print(f"length of content: {filtered_df[['word_count']]}")
    return positive_df
```

```python
# Function to get 3 random negative examples, allowing for >1 search term using ['...', '...']
def improved_neg_search_query(query, query_number, search_terms, search_columns, df, target_column):
    pattern = '|'.join(search_terms)
    filtered_df = df[~df[search_columns].str.contains(pattern, case=False, na=False)]
    random_rows = filtered_df.sample(n=3, random_state=42)
    random_rows['word_count'] = filtered_df[target_column].str.split().apply(len)

    negative_df = random_rows.copy()
    negative_df['query'] = query
    negative_df['query_number'] = query_number
    negative_df['sample'] = 'negative'

    print(random_rows[['Article Links', target_column, 'word_count']])
    print(f"length of content: {random_rows[['word_count']]}")
    return negative_df
```
